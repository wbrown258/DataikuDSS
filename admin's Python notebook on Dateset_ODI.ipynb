```python
import dataiku
from dataiku import pandasutils as pdu
import pandas as pd
import numpy as np
```

```python
# Read the dataset as a Pandas dataframe in memory
# Note: here, we only read the first 100K rows. Other sampling options are available
dataset_Dateset_ODI = dataiku.Dataset("Dateset_ODI")
df = dataset_Dateset_ODI.get_dataframe(limit=100000)
```

```python
# Get some simple descriptive statistics
pdu.audit(df)
```

```python
## Name the ground on which the maximum number of matches were played?
df['Ground'].value_counts()
```

```python
##Which team won the maximum number of Matches ?
df['Winner'].value_counts()
```

```python
#Which team won maximum matches on Ground ‘ Sharjah’ ?
df1= df[df['Ground']=='Sharjah']
df1['Winner'].value_counts()
```

```python
#Which country won max average no. of matches - i.e.max( total wins / total matches played)

total_no_matches_played = df['Team 1'].value_counts() + df['Team 2'].value_counts()
#print(total_no_matches_played)
Average = df['Winner'].value_counts() / total_no_matches_played
#print(Average)
max_country = Average.nlargest(1)
print(max_country)
```

```python
#What was the date on which exactly 6 matches were played?

df['Match Date'].value_counts()
```

```python
#Which ground did Australia win the most number of matches?

df1 = df[df['Winner']== 'Australia']
df1['Ground'].value_counts()
```

```python
#Which team won the most matches on Lord’s Ground?

df1 = df[df['Ground'] == "Lord's"]
df1['Winner'].value_counts()
```

```python
#How many times team 2 won?

df['Team 2'] == df['Winner'].sum()
```

```python
#From all the winning matches in Team 2, which team won the most number of matches?

df1 = df[df['Team 2'] == df['Winner']]
df1['Winner'].value_counts()
```

```python
#Which teams played on Feb 4, 2007 ?

df1 = df[df['Match Date']== 'Feb 4, 2007']
print(df1)
```
